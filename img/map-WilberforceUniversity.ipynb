```python
# import python libraries
import pandas as pd 
import hvplot.pandas
import holoviews as hv
import osmnx as osm
```

```python
# Search for boundary using OSM

 

wu_boundary = osm.geocode_to_gdf(

                        "W128954737",

                        by_osmid=True

)

 

wu_boundary
```

```python
wu_boundary.plot()
```

```python
# Make an interactive plot

wu_map = wu_boundary.hvplot(

    geo=True, tiles='EsriImagery'

)

 

# Save the map as a file to put on the web (OPTIONAL)

hv.save(wu_map, 'wu_map.html')

 

# Display the map

wu_map
```
